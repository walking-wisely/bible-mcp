# bible-mcp seed

Builds `bible-web-nomic.db` and uploads it to Cloudflare R2.

**Before running:**
1. Enable GPU: *Session options → Accelerator → GPU T4 x2*
2. Add R2 credentials under *Add-ons → Secrets*:
   - `R2_ACCOUNT_ID`
   - `R2_ACCESS_KEY_ID`
   - `R2_SECRET_ACCESS_KEY`
   - `R2_BUCKET`
3. Hit **Run All**.

```python
# Load R2 credentials from Kaggle Secrets into environment variables
import os
from kaggle_secrets import UserSecretsClient

secrets = UserSecretsClient()
for key in ["R2_ACCOUNT_ID", "R2_ACCESS_KEY_ID", "R2_SECRET_ACCESS_KEY", "R2_BUCKET"]:
    os.environ[key] = secrets.get_secret(key)
```
